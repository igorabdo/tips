
# --------------------- PART 1: Parameters ---------------------

```python
# Target Delta Table Variables
datalakehouse_source = ''
datalakehouse_target = 'controller'
target_table_name = 'controller'

# Datalake name
delta_table_name_target = f'{datalakehouse_target}.{target_table_name}'

pk_columns = [""]
optimize_columns = ""
partition_columns = 'N/A'
replace_column = 'N/A'
merge_columns = ""
table_type = 'managed'

write_table_model = 'overwrite' ## "overwrite" ou "replace_where" ou "merge"
apply_distinct = "False" ## "False"
apply_hash = "False" ## "False"
apply_sk = False
columns_sk = ''

```

# --------------------- PART 2: Create Dataframe with the Data --------------------

```python
data = [
    (1, 1, "Person", "AddressType", "SELECT * FROM Person.AddressType", "Person", "Person_AddressType.parquet"),
    (2, 1, "Person", "BusinessEntity", "SELECT * FROM Person.BusinessEntity", "Person", "Person_BusinessEntity.parquet"),
    (3, 1, "Person", "BusinessEntityAddress", "SELECT * FROM Person.BusinessEntityAddress", "Person", "Person_BusinessEntityAddress.parquet"),
    (70, 1, "Person", "Address", "SELECT AddressID, AddressLine1, AddressLine2, City, StateProvinceID, PostalCode, rowguid, ModifiedDate FROM Person.Address", "Person", "Person_Address.parquet")
]

columns = ["Id", "Version", "SchemaName", "TableName", "Query", "TargetSchema", "FileName"]

df = spark.createDataFrame(data, columns)
```


# --------------------- PART 3: Check if the table exists ---------------------

```python
if not spark.catalog.tableExists(delta_table_name_target):
    # Create table if it does not exist
    df.write.format("delta").saveAsTable(delta_table_name_target)
    print(f"Table {delta_table_name_target} created.")
else:
    print(f"Table {delta_table_name_target} is exist.")
```

# --------------------- PART 4: Write DataFrame to Table ---------------------

```python
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(delta_table_name_target)
print(f"Data overwritten in the table {delta_table_name_target}.")
```
